```python
import numpy as np
```

```python
folder_names = ['OOD/odin_scores/', 'OOD/openmax_scores/', 'OOD/softmax_scores/']
titles = ['Tiny ImageNet', 'SVHN', 'FakeData']
odin_names = [
    ['odin_cifar_tiny_DenseNet_augm', 'odin_cifar_tiny_VGG_augm', 'odin_cifar_tiny_WRN28_augm', 'odin_cifar_tiny_WRN40_augm'],
    ['odin_cifar_svhn_DenseNet_augm', 'odin_cifar_svhn_VGG_augm', 'odin_cifar_svhn_WRN28_augm', 'odin_cifar_svhn_WRN40_augm'],
    ['odin_cifar_fake_DenseNet_augm', 'odin_cifar_fake_VGG_augm', 'odin_cifar_fake_WRN28_augm', 'odin_cifar_fake_WRN40_augm']
]

openmax_names = [
    ['openmax_cifar_tiny_DenseNet_augm', 'openmax_cifar_tiny_VGG_augm', 'openmax_cifar_tiny_WRN28_augm', 'openmax_cifar_tiny_WRN40_augm'],
    ['openmax_cifar_svhn_DenseNet_augm', 'openmax_cifar_svhn_VGG_augm', 'openmax_cifar_svhn_WRN28_augm', 'openmax_cifar_svhn_WRN40_augm'],
    ['openmax_cifar_fake_DenseNet_augm', 'openmax_cifar_fake_VGG_augm', 'openmax_cifar_fake_WRN28_augm', 'openmax_cifar_fake_WRN40_augm']
]

softmax_names = [
    ['softmax_cifar_tiny_DenseNet_augm', 'softmax_cifar_tiny_VGG_augm', 'softmax_cifar_tiny_WRN28_augm', 'softmax_cifar_tiny_WRN40_augm'],
    ['softmax_cifar_svhn_DenseNet_augm', 'softmax_cifar_svhn_VGG_augm', 'softmax_cifar_svhn_WRN28_augm', 'softmax_cifar_svhn_WRN40_augm'],
    ['softmax_cifar_fake_DenseNet_augm', 'softmax_cifar_fake_VGG_augm', 'softmax_cifar_fake_WRN28_augm', 'softmax_cifar_fake_WRN40_augm']
]

training_runs = ['DenseNet_augm-results.txt', 'VGG_augm-results.txt','WRN28_augm-results.txt','WRN40_augm-results.txt']

```

```python
# Print the IST metrics as defined in AI Test. 

def print_ist_metrics(j): 
    for i in range(0,3): 
        for j in range(j,j+1):
            ### Print ODIN results
            training_results = np.loadtxt('./Training results/' + training_runs[j], skiprows=1)
            odin_ood_scores = np.loadtxt(folder_names[0] + odin_names[i][j] + '.txt', delimiter=',')
            odin_epoch = odin_ood_scores[1,0].astype('int') - 1
            odin_results = odin_ood_scores[1,:]*100


            ### Print OPENMAX results
            training_results = np.loadtxt('./Training results/' + training_runs[j], skiprows=1)
            om_ood_scores = np.loadtxt(folder_names[1] + openmax_names[i][j] + '.txt', delimiter=',')
            om_epoch = om_ood_scores[1,0].astype('int') - 1
            om_results = om_ood_scores[1,:]*100

            ### Print Baseline (Softmax) results
            training_results = np.loadtxt('./Training results/' + training_runs[j], skiprows=1)
            bl_ood_scores = np.loadtxt(folder_names[2] + softmax_names[i][j] + '.txt', delimiter=',')
            bl_epoch = bl_ood_scores[1,0].astype('int') - 1
            bl_results = bl_ood_scores[1,:]*100

            # print(training_runs[j], odin_names[i][j], openmax_names[i][j], softmax_names[i][j])
            for k in range(1,5): 
                if k != 4: 
                    print("{:2.2f} / {:2.2f} / {:2.2f} & ".format(odin_results[k], om_results[k], bl_results[k]), end='')
                else: 
                    print("{:2.2f} / {:2.2f} / {:2.2f} \\\\ ".format(odin_results[k], om_results[k], bl_results[k]), end='')
            print("")
    print('\\hline')

def print_names(j):
    for i in range(0,3): 
        for j in range(j,j+1):
            print(training_runs[j], odin_names[i][j], openmax_names[i][j], softmax_names[i][j])
```

```python
print_ist_metrics(0)
print_ist_metrics(1)
print_ist_metrics(2)
print_ist_metrics(3)

print_names(0)
print_names(1)
print_names(2)
print_names(3)
```

```python
# Print the Pearson coefficients for all the combinations 

from scipy.stats import pearsonr

for i in range(0,3): # For all 3 O.O.D datasets
    all_accuracy_scores = [] 
    all_auroc_scores = [] 
    all_fpr_scores = [] 
    all_cbpl_scores = [] 
    all_cov10_scores = [] 
    for j in range(4): # For all 4 models 
        ### Print ODIN results
        training_results = np.loadtxt('./Training results/' + training_runs[j], skiprows=1)
        ood_scores = np.loadtxt(folder_names[0] + odin_names[i][j] + '.txt', delimiter=',')
        epochs = ood_scores[:,0].astype('int') - 1
        accuracy = training_results[epochs,1]
        auroc = ood_scores[:,1]
        fpr95 = ood_scores[:,2]
        cbpl = ood_scores[:,3]
        cov10 = ood_scores[:,4]
        [all_accuracy_scores.append(x) for x in accuracy]
        [all_auroc_scores.append(x) for x in auroc]
        [all_fpr_scores.append(x) for x in fpr95]
        [all_cbpl_scores.append(x) for x in cbpl]
        [all_cov10_scores.append(x) for x in cov10]
        
        ### Print OPENMAX results
        training_results = np.loadtxt('./Training results/' + training_runs[j], skiprows=1)
        ood_scores = np.loadtxt(folder_names[1] + openmax_names[i][j] + '.txt', delimiter=',')
        epochs = ood_scores[:,0].astype('int') - 1
        accuracy = training_results[epochs,1]
        auroc = ood_scores[:,1]
        fpr95 = ood_scores[:,2]
        cbpl = ood_scores[:,3]
        cov10 = ood_scores[:,4]
        [all_accuracy_scores.append(x) for x in accuracy]
        [all_auroc_scores.append(x) for x in auroc]
        [all_fpr_scores.append(x) for x in fpr95]
        [all_cbpl_scores.append(x) for x in cbpl]
        [all_cov10_scores.append(x) for x in cov10]

        
        ### Print Baseline (Softmax) results
        training_results = np.loadtxt('./Training results/' + training_runs[j], skiprows=1)
        ood_scores = np.loadtxt(folder_names[2] + softmax_names[i][j] + '.txt', delimiter=',')
        epochs = ood_scores[:,0].astype('int') - 1
        accuracy = training_results[epochs,1]
        auroc = ood_scores[:,1]
        fpr95 = ood_scores[:,2]
        cbpl = ood_scores[:,3]
        cov10 = ood_scores[:,4]
        [all_accuracy_scores.append(x) for x in accuracy]
        [all_auroc_scores.append(x) for x in auroc]
        [all_fpr_scores.append(x) for x in fpr95]
        [all_cbpl_scores.append(x) for x in cbpl]
        [all_cov10_scores.append(x) for x in cov10]

    corr_auroc = pearsonr(all_accuracy_scores, all_auroc_scores)
    corr_fpr95 = pearsonr(all_accuracy_scores, all_fpr_scores)
    corr_cbpl = pearsonr(all_accuracy_scores, all_cbpl_scores)
    corr_cov10 = pearsonr(all_accuracy_scores, all_cov10_scores)
    
    print("{} accuracy-auroc corr: {}".format(titles[i], corr_auroc))
    print("{} accuracy-FPR95 corr: {}".format(titles[i], corr_fpr95))
    print("{} accuracy-CBPL corr: {}".format(titles[i], corr_cbpl))
    print("{} accuracy-cov10 corr: {}".format(titles[i], corr_cov10))
```
